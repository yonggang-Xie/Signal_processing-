```python
def random_slicing(paths,filenames,origsr,targetsr,time,t_start,mode):
    '''
    This function cuts audio files into pieces of user-input duration.This allows random or fixed starting point start.
    set the mode to random to produce random starting point.
    
    Inputs
    =====
    
    path  : list of string
          A list of strings, where each string is the path to an audio file.
    
    filenames : list of string
           A list of strings, where each string is the path to an audio file.
    
    origsr   : int
           original sample rate
    
    targetsr   : int
           target sample rate
    
    time: int
           the duration you want to slice
    
    t_start : int
            the starting point of your slice in seconds.
            t_start is a don't-care value if mode is set to 'random'.
    
    mode   : string
           set the function to 'random',otherwise it is nonrandom.
    
    outputs
    =====
    
    write the sliced files to the assigned location.
    '''
        
    data , sr= sf.read(paths)
    
    np.random.seed(2020)
            
    d_resample = librosa.resample(np.transpose(data),orig_sr=origsr,target_sr=targetsr)
            
    d_transpose = np.transpose(d_resample)
    
    t_random = random.randint(0,targetsr*time)
    
    if (mode=='random'):
        
        
        t_end = t_random + targetsr*time
        
        new = d_transpose[t_random : t_end]
    
    else :   
        
        t_end = t_start + time
        
        new = d_transpose[t_start * targetsr: t_end * targetsr]   
    
            
    sf.write(filenames,new,targetsr)
    
```
